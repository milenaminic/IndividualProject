```python
path = 'C:/Users/jet769/Desktop/Individual project/DA Case_Q12023/Raw data/TransactionsData_csv.csv'
df_transaction = pd.read_csv (path)
df_transaction
```

```python
df_transaction.replace(" ", np.nan, inplace = True)
df_transaction.head(5)
```

```python
print(df_transaction["PAYMENT"].value_counts())
```

```python
#replaceing all NaN values where payment is done with any payments metod (in original table of transaction we had empty rows for "PURCHASE")
condition = (df_transaction['PURCHASE']== 'Nan') & (
    (df_transaction['PAYMENT'] == 'PAYPAL') | 
    (df_transaction['PAYMENT'] == 'APPLE PAY ') | 
    (df_transaction['PAYMENT'] == 'VISA/MASTERCARD') | 
    (df_transaction['PAYMENT'] == 'AMEX')
    )
df_transaction['PURCHASE'] = np.where(condition, 'YES', df_transaction['PURCHASE'])
```

```python
#add a column 'AMOUNT before discount' and calculate it as QTY * PURCHASEPRICE

merged_df = df_transaction.merge(df_item, left_on='ITEM', right_on='ITEM', how='left')

merged_df['AMOUNT before %'] = merged_df['QTY'] * merged_df['SALEPRICE']

merged_df.head(5)
```

```python
#add a column 'amount after discount'
merged_df['AMOUNT after %'] = merged_df['AMOUNT before %'] * (1 - merged_df['DISCOUNT'] / 100)
merged_df.head(5)
```

-CREATING CALENDAR TABLE for analysis in Qlik-

```python
# Create a list of dates for the year 2018
start_date = pd.to_datetime('01/01/2018')
end_date = pd.to_datetime('31/12/2018')
date_range = pd.date_range(start_date, end_date)

# Create a DataFrame with Date, Month (full name), and Month Abbreviation columns
df = pd.DataFrame({
    'Date': date_range,
    'Month': date_range.strftime('%B'),  # Full month names
    'MonthAbbr': date_range.strftime('%b'),  # Month abbreviations
    'Weekday': date_range.strftime('%A'),
    'Year': date_range.strftime('%Y')    # Year
    
})
df['MonthYear'] = df['Date'].dt.strftime('%m-%Y')
# Define a dictionary to map month abbreviations to numeric values
month_mapping = {
    'Jan': 1, 'Feb': 2, 'Mar': 3, 'Apr': 4, 'May': 5, 'Jun': 6,
    'Jul': 7, 'Aug': 8, 'Sep': 9, 'Oct': 10, 'Nov': 11, 'Dec': 12
}

# Add the 'MonthNbr' column based on 'MonthAbbr' using the map function
df['MonthNbr'] = df['MonthAbbr'].map(month_mapping)

# Display the DataFrame
print(df)




```
